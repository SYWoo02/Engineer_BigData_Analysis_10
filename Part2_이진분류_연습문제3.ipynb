# 이진분류 연습문제

## Section 3 신용카드 신청자의 미래 신용 예측
### 신용카드 신청자의 채무 불이행을 예측하시오.
### 학습용 데이터(train)를 이용해 새 일자리를 찾을지 예측하는 모델을 만든 후 이를 평가용 데이터(test)에 적용해 얻은 예측값을 다음과 같은 형식의 CSV 파일로 생성하시오
* 예측한 칼럼: STATUS(0: 채무 이행, 1: 채무 불이행)
* pred: 예측값
* 제출 파일명: 'result.csv'
* 제출한 모델의 성능은 f1 평가지표에 따라 채점한다.

```python
# 라이브러리 및 데이터 불러오기
import pandas as pd

train = pd.read_csv("creditcard_train.csv")
test = pd.read_csv("creditcard_test.csv")
```

```python
# EDA
print("===== Train 데이터 확인 =====")
print(train.head(5))
print("\n")

print("===== Test 데이터 확인 =====")
print(test.head(5))
print("\n")

print("===== 데이터 타입 확인 =====")
print(train.info(5))
print("\n")

print("===== Target의 분포 확인 =====")
print(train['STATUS'].value_counts())
print("\n")

print("===== Train 데이터 결측치 확인 =====")
print(train.isnull().sum())
print("\n")

print("===== Test 데이터 결측치 확인 =====")
print(test.isnull().sum())
print("\n")
```

```python
# train과 test 범주형 데이터 비교
cols = train.select_dtypes(include='object').columns
for col in cols:
  set_train = set(train[col])
  set_test = set(test[col])
  same = (set_train == set_test)
  if same:
    print(col,"\t카테고리 동일함")
  else:
    print(col,"\t카테고리 동일하지 않음")
```

```python
# 결측치 처리
train.dropna(subset=['OCCUPATION_TYPE'], inplace=True)

# Train에서 target 분리
target = train.pop('STATUS')
```

```python
# 원-핫 인코딩
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
# Validation 데이터 분할
from sklearn.model_selection import train_test_split
X_train, X_val, y_train, y_val = train_test_split(train, target, test_size=0.3, random_state=0)

# 머신러닝 학습 및 평가
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0)
rf.fit(X_train, y_train)
pred = rf.predict(X_val)

from sklearn.metrics import f1_score
score = f1_score(y_val, pred)
print("f1:", score)
```

```python
# 예측 및 결과 파일 생성
pred = rf.predict(test)
submit = pd.DataFrame({'pred':pred})
submit.to_csv("result.csv", index=False)

# 결과 파일 확인
submit = pd.read_csv("result.csv")
submit
```
