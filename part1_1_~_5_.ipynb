# Section 01 필터링, 최솟값, 중앙값

```python
import pandas as pd
df1 = pd.read_csv('type1_data1.csv')
df1.head()
```

## 1. 'f5' 컬럼이 0이 아닌 데이터(행)을 구하시오.

```python
df_f5_non_zero = df1[df1['f5'] != 0]
df_f5_non_zero['f5']
```

## 2. 앞에서 구한 데이터에 'views' 컬럼 결측치를 'views' 컬럼의 최솟값으로 채워주세요.

```python
## 'views' 컬럼에 결측치는 몇 개?
df_f5_non_zero.isnull().sum()
```

```python
min = df_f5_non_zero['views'].min()
df_f5_non_zero['views'].fillna(min, inplace=True)

## 'views' 컬럼에 결측치가 없음을 확인!
df_f5_non_zero.isnull().sum()
```

## 3. 그리고 'views' 컬럼의 중앙값을 계산해 정수로 구하시오.

```python
median = df_f5_non_zero['views'].median()
print(int(median))
```

# Section 03 파생변수, 정렬, 인덱싱

```python
import pandas as pd
df3 = pd.read_csv('type1_data1.csv')
df3.head()
```

## 1. 결측치가 있는 데이터(행)을 제거하시오.

```python
df_3 = df3
df_3.dropna(inplace=True)
df_3.isnull().sum()
```

## 2. 'views' 컬럼을 'f1' 컬럼으로 니눈 값을 새로운 컬럼으로 추가하시오.

```python
df_3['new'] = 0
df_3['new'] = df_3['views'] / df_3['f1']
df_3.head()
```

## 3. 새로운 컬럼 값 중 가장 큰 값을 가진 행의 age를 정수로 구하시오.

```python
# max = df['new'].max()
# idx = df[df['new'] == max]['age']
# int(idx)

# 내림차순 정렬
df_3_ = df_3.sort_values('new', ascending=False)
int(df_3_.iloc[0,1])
```

# Section 04 값 변경, 정렬, 합계

```python
import pandas as pd
df4 = pd.read_csv('type1_data1.csv')
df4.head()
```

## 1. 'views' 컬럼의 결측 데이터를 0으로 대체하시오.

```python
df_4 = df4
df_4 = df_4.fillna(0)
df_4.isnull().sum()
```

## 2. 'views' 컬럼에서 상위 10번째 값을 구하시오.

```python
df_4 = df_4.sort_values('views', ascending=False)
df_4['views'].head(10)
```

```python
views_min = df_4.iloc[:10]['views'].min()
views_min
```

## 3. 'views' 컬럼에서 상위 10개의 값을 상위 10번째 값으로 대체하시오.

```python
df_4.iloc[:10, -1] = views_min
df_4.head(10)
```

## 4. 'views' 컬럼의 전체 합을 정수로 구하시오.

```python
sum = df_4['views'].sum()
print(int(sum))
```

# Section 5 문자열 슬라이싱, 파생변수, 평균값

```python
import pandas as pd
df5 = pd.read_csv('type1_data1.csv')
df5.head()
```

## 1. 'f4' 컬럼 데이터에 'FJ'가 포함된 데이터를 찾으시오.

```python
df_5 = df5
cond = df_5['f4'].str.contains('FJ')
df_5 = df_5[cond]
df_5['f4']
```

## 2. 찾은 데이터 중에서 'f2' 컬럼의 평균값을 구하시오. (반올림 후 소수 둘째 자리까지 계산)

```python
mean = df_5['f2'].mean()
print(round(mean,2))
```
