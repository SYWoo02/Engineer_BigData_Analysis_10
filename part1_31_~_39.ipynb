# Section 31 문자열, 형 변환

# Section 33 데이터프레임 재구조화

```python
df33 = pd.read_csv("school_data.csv")
df33
```

## 1. 과목에 상관없이 점수가 가장 작은 점수 25개를 찾으시오.

```python
# 이름을 유지하면서 수학, 영어, 국어 형태 변경 (열 -> 행)
melted_df33 = df33.melt(id_vars=['이름'], value_vars=['수학','영어','국어'])
melted_df33
```

# Section 34 데이터 합치기(concat)

```python
df34 = pd.read_csv("school_data.csv")
df34_science = pd.read_csv("school_data_science.csv")
df34_science
```

## 1. 주어진 두 csv파일을 학생 순서를 기준으로 병합하시오. (단, 두 파일의 학생 순서는 동일하다.)

```python
# 두 파일 병합 (열로 ; axis = 1)
new_df34 = pd.concat([df34, df34_science], axis=1)
new_df34
```

## 2. 학생별로 수학, 영어, 국어, 과학 점수의 평균을 구하시오.

```python
new_df34['과목 평균'] = (new_df34['수학'] + new_df34['영어'] + new_df34['국어'] + new_df34['과학']) / 4
new_df34['과목 평균']
```

## 3. 평균 점수가 60점 이상인 인원 수를 계산하시오.

```python
count = (new_df34['과목 평균'] >= 60).sum()
print(count)
```

# Section 35 데이터합치기 (merge)

```python
df35 = pd.read_csv("school_data.csv")
df35_social = pd.read_csv("school_data_social.csv")
df35_social
```

## 1. 두 파일을 '이름을 기준으로 합치시오.

## 2. 영어교사가 장선생이면서 사회교사가 오선생인 학생들을 필터링하시오.

## 3. 필터링된 학생들의 수학 점수를 모두 더한 후 정수로 구하시오.

```python
total_math = filtered_new_df35['수학'].sum()
print(total_math)
```

# Section 36 조건별 변환(transform)

```python
df36 = pd.read_csv("sales.csv")
df36
```

## 1. 결측된 판매금액은 해당 지역의 평균 판매금액으로 결측값을 대체하시오.

```python
df36.isnull().sum()
```

```python
df36  = df36.fillna(df36['판매금액'].mean())
df36.isnull().sum()
```

## 2. 각 거래마다 "판매금액"과 해당 지역의 평균 판매금액의 "차이"를 절대값으로 구하시오.

```python
# 해당 지역의 평균 판매금액
# pd.transform(): 그룹별로 계산된 결과를 원래 데이터의 형태(행 수)를 유지하면서 반환할 때 사용
df36['지역평균'] = df36.groupby(by='지역코드')['판매금액'].transform('mean')

# 각 거래별 판매 금액과 해당 지역의 평균 판매금액의 차이 (절대값)
df36['차이'] = abs(df36['판매금액'] - df36['지역평균'])

df36['차이']
```

## 3. 각 지역에서 차이값의 평균을 구한 후, 이 값이 가장 큰 지역의 지역코드를 구하시오.

```python
# 지역별 차이값의 평균을 구함 -> idxmax()로 그 값이 가장 큰 지역의 지역코드 출력
result = df36.groupby(by='지역코드')['차이'].mean().idxmax()
print(result)
```

# Section 37 재구조화(unstack), 맵핑

```python
df37 = pd.read_csv("store_sales.csv")
df37
```

## 1. 각 행에서 판매수량과 단가를 이용하여 매출액을 계산하시오.

```python
df37['매출액'] = df37['판매수량'] * df37['단가']
df37['매출액']
```

## 2. 요일을 평일과 주말로 구분하고, 매장별 평일과 주말 매출액 합계를 구하시오.

```python
# 요일을 평일과 주말로 구분 (map() 이용)
day_mapping = {'월':'평일', '화':'평일', '수':'평일', '목':'평일', '금':'평일', '토':'주말', '일':'주말'}
df37['구분'] = df37['요일'].map(day_mapping)

# 매장별 평일과 주말 매출액 합계 계산
sales_by_days_branch = df37.groupby(by=['매장코드','구분'])['매출액'].sum()
sales_by_days_branch
```

## 3. 매장별 평일과 주말 매출액 차이를 절대값으로 구하시오. 이후, 모든 매장 중 가장 큰 절대값 차이를 찾으시오.

```python
# unstack(): 데이터프레임의 행 인덱스를 열 인덱스로 변환
df37_ = sales_by_days_branch.unstack()

df37_['평일-주말 간 매출차이'] = abs(df37_['주말'] - df37_['평일'])
max = df37_['평일-주말 간 매출차이'].max()
print(max)
```

# Section 38 피벗테이블

```python
df38 = pd.read_csv("region_sales.csv")
df38
```

## 1. 각 Region과 Channel 조합별로 제품(Product) A, B의 총 판매액을 계산하시오.

```python
# pd.pivot_table(
#     data,
#     values=None,         # 집계 대상 열
#     index=None,          # 행 인덱스로 사용할 열
#     columns=None,        # 열 인덱스로 사용할 열
#     aggfunc='mean',      # 집계 함수: 기본은 평균(mean)
#     fill_value=None      # 결측값 대체 값
# )
```

```python
df38_ = pd.pivot_table(
    df38,
    values='Sales',         # 집계 대상 열
    index=['Region','Channel'],          # 행 인덱스로 사용할 열
    columns='Product',        # 열 인덱스로 사용할 열
    aggfunc='sum',      # 집계 함수: 기본은 평균(mean)
    fill_value=None      # 결측값 대체 값
)
df38_
```

## 2. 제품 A의 매출 비율(A비율)을 구하시오. A비율 = (제품 A 판매액) / (제품 A 판매액 + 제품 B 판매액)

```python
df38_['A_rate']= df38_['A'] / (df38_['A'] + df38_['B'])
df38_['B_rate']= df38_['B'] / (df38_['A'] + df38_['B'])
df38_
```

## 3. A비율 중 최댓값을 찾아 반올림하여 소수 둘째 자리까지 구하시오.

```python
max = df38_['A_rate'].max()
print(round(max,2))
```

## 1. 지역(Region)과 월(Jan, Feb, Mar)별 매출(Sales) 합계를 구하시오.

```python
# pd.melt(
#     frame,                         # 변환할 데이터프레임
#     id_vars=None,                  # 그대로 유지할 열
#     value_vars=None,               # 녹일 열들 (default: 나머지 전부)
#     var_name='variable',           # 녹인 열 이름
#     value_name='value'             # 녹인 값 이름
# )
# 열 -> 행
```

```python
# Region열만 남기고 월(Jan, Feb, Mar)열들을 melt(열 -> 행)
melted_df39 = pd.melt(df39,id_vars='Region')
# melted_df39

# Region과 variable(월) 별 value(매출) 합계 계산
total_Sales = melted_df39.groupby(by=['Region','variable'])['value'].sum()
total_Sales
```

## 2. 위에서 구한 결과 중, 매출 합계(Sales)가 1400을 초과하는 경우가 몇 건인지 구하시오.

```python
print(total_Sales[total_Sales>1400].value_counts().sum())
```
