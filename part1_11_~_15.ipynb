# Section 11 슬라이싱, 사분위수, 결측치 제거

```python
import pandas as pd
df11 = pd.read_csv('type1_data1.csv')
df11.head()
```

## 1. 데이터에서 결측치가 있는 데이터(행)을 모두 제거하시오.

```python
df_11 = df11.dropna()
df_11
```

## 2. 결측치가 제거된 데이터를 사용해 앞에서부터 70% 데이터를 구하시오. (단, 데이터 70% 지점의 값이 소수점으로 계산될 경우 소수점 이하는 버림)

```python
df_11_70 = df_11.iloc[:int(len(df_11) * 0.7)]
df_11_70
```

## 3. 앞에서 구한 70% 데이터 중 'views' 칼럼의 3사분위수에서 1사분위수를 뺀 값을 정수로 구하시오.

```python
result = df_11_70['views'].quantile(0.75) - df_11_70['views'].quantile(0.25)
print(int(result))
```

# Section 12 결측치 처리, 최빈값, 데이터 개수

```python
import pandas as pd
df12 = pd.read_csv('type1_data1.csv')
```

## 1. 결측치가 가장 많은 두 컬럼을 찾으시오.

```python
missing = df12.isnull().sum()
missing = missing.sort_values(ascending=False)
print(missing.index[:2])
```

## 2. 첫 번째로 결측치가 많은 컬럼에서 결측치가 있는 데이터(행)를 삭제하시오.

```python
df12 = df12.dropna(subset = ['f1'])
```

## 3. 두 번째로 결측치가 많은 컬럼을 최빈값으로 대체하시오.

```python
freq = df12['f3'].mode()[0]
df12['f3'] = df12['f3'].fillna(freq)
```

## 4. 'f3' 컬럼의 'gold' 값을 가진 데이터의 수를 정수형으로 구하시오.

```python
cond = df12['f3'] == 'gold'
print(len(df12[cond]))
```

# Section 13 결측 데이터 찾기, 필터링, 평균값

## 2. 선택된 데이터에서 'age' 컬럼의 평균값을 구하시오. (반올림 후 소수 첫째 자리까지 계산)

# Section 14 중복 데이터 제거, 값 변경, 데이터 개수

```python
import pandas as pd
import numpy as np
df14 = pd.read_csv('type1_data1.csv')
```

## 1. 중복 데이터를 제거하시오.

```python
df_14 = df14.drop_duplicates()
df_14.head(5)
```

## 2. 'f3' 컬럼의 결측치는 0, 'silver'는 1, 'gold'는 2, 'vip'는 3으로 변환하시오.

```python
# 방법 1
df_14['f3'] = df_14['f3'].replace(np.nan,0).replace('silver',1).replace('gold',2).replace('vip',3)
df_14.head(5)
```

```python
# 방법 2
df_14 = df_14.fillna(0)
df_14['f3'] = df_14['f3'].map({'silver':1, 'gold' : 2, 'vip': 3})
df_14.head(5)
```

## 3. 변환된 'f3' 컬럼의 총합을 정수형으로 구하시오.

```python
sum = df_14['f3'].sum()
print(int(sum))
```

# Section 15 컬럼 삭제, 행 단위 합계, 필터링

```python
import pandas as pd
df15 = pd.read_csv('type1_data1.csv')
```

## 1. 주어진 데이터에서 문자 자료형 컬럼을 삭제하시오.

```python
df_15 = df15.select_dtypes(exclude='object')
df_15.head(5)
```

## 2. 숫자 자료형 컬럼의 결측치를 0으로 대체하시오.

```python
df_15.isnull().sum()
```

```python
df_15 = df_15.fillna(0)
df_15.head(5)
```

## 3. 각 행의 합이 3000보다 큰 값의 개수를 정수로 구하시오.
(각 행의 합: 'age' + 'f1' + 'f2' + 'f5' + 'views')

```python
df_15 = df_15.T
df_15
```

```python
# 각 행의 합 계산
row_sums = df_15.sum(axis=1)

# 합이 3000보다 큰 행의 개수
count_over_3000 = (row_sums > 3000).sum()

# 결과 출력
print(int(count_over_3000))
```
