# Section 26 날짜와 시간 정보 변환, 비율

```python
import pandas as pd
df26 = pd.read_csv("delivery_time.csv")
df26
```

## 1. 주문이 가장 많이 발생한 연-월을 찾으시오.

```python
# 주문시간 열을 datetime 형식으로 변환
df26['주문시간'] = pd.to_datetime(df26['주문시간'])

# 주문시간 열을 월 단위로 변환
df26['주문월'] = df26['주문시간'].dt.to_period('M')

# 주문이 가장 많이 발생한 연-월 찾기
year_month = df26['주문월'].value_counts().idxmax()
year_month
```

## 2. 해당 연-월에 '배고팡' 앱을 통한 주문 중 '앱결제'로 결제된 주문의 비율을 계산하시오. (반올림 후 소수 둘째 자리까지 계산)

```python
# 전체 중 해당 연-월에 '배고팡'앱을 통한 주문들만 필터링
cond1 = df26['주문월'] == year_month
cond2 = df26['앱종류'] == '배고팡'
filtered_df26 = df26[cond1 & cond2]

# 필터링된 데이터들 중 '앱결제'로 결제된 주문만 필터링
cond3 = filtered_df26['결제종류'] == '앱결제'

# 비율 구하기
result = len(filtered_df26[cond3]) / len(filtered_df26)
print(round(result,2))
```

# Section 27 시간 범위, 속도(km/h)

```python
import pandas as pd
df27 = pd.read_csv("delivery_time.csv")
df27
```

## 1. 점심시간(10시부터 13시 전까지)에 주문된 배달 데이터를 찾으시오.
* 배달시간 = 실제도착시간 - 주문시간
* 속도(km/h) = 거리(km) / 시간(h)

```python
df27['실제도착시간'] = pd.to_datetime(df27['실제도착시간'])
df27['주문시간'] = pd.to_datetime(df27['주문시간'])

# 점심시간 주문 데이터 찾기
# 1. 주문시간 열에서 시간(시) 추출
df27['주문시간_hour'] = df27['주문시간'].dt.hour

# 2. 점심시간(10시부터 13시 전)이라는 조건 설정
cond1 = df27['주문시간_hour'] >= 10
cond2 = df27['주문시간_hour'] < 13
df27_lunch = df27[cond1 & cond2].copy()
df27_lunch
```

## 2. 점심시간 주문 건 중 과속(평균 속도가 50km/h 이상)하는 주문 수를 정수로 구하시오.


*   배달시간 = 실제도착시간 - 주문시간
*   속도(km/h) = 거리(km) / 시간(h)



```python
# 기간을 초로 변환 total_seconds()
# print(diff.dt.total_seconds()) #초
# print(diff.dt.total_seconds()/60) #분
# print(diff.dt.total_seconds()/60/60) #시간
# print(diff.dt.total_seconds()/60/60/24) #일
```

```python
df27_lunch['배달시간'] = (df27_lunch['실제도착시간'] - df27_lunch['주문시간']).dt.total_seconds() /60 /60
df27_lunch['배달속도'] = df27_lunch['거리'] / df27_lunch['배달시간']
print("점심시간 주문 건 중 과속을 하는 주문 수: ", sum(df27_lunch['배달속도'] >= 50))
```

# Section 28 날짜와 시간, 문자열

```python
import pandas as pd
df28 = pd.read_csv("delivery_time.csv")
df28
```

## 1. 연도와 월을 기준으로 주문 수를 집계하시오.

```python
# Section 26과 유사

# 주문시간 열을 datetime 형식으로 변환
df28['주문시간'] = pd.to_datetime(df28['주문시간'])

# 주문시간 열을 월 단위로 변환
df28['주문월'] = df28['주문시간'].dt.to_period('M')

# 주문 월을 기준으로 주문 횟수 계산
order_count = df28.groupby('주문월').size()
order_count
```

## 2. 가장 많은 주문이 있었던 연도와 월을 예시와 같은 형식으로 숫자로만 구하시오. (예: 2024년 2월인 경우 '202402', 2024년 10월인 경우 '202410')

```python
# 가장 주문이 많았던 달의 인덱스 추출
max_order_month = order_count.idxmax()

# 예시같은 형식으로 출력 (하이픈 제거)
print(str(max_order_month).replace("-",""))
```

# Section 29 함수, 월별 집계

```python
import pandas as pd
df29 = pd.read_csv("delivery_time.csv")
df29
```

## 1. 배달료 계산 기준표에 따라 각 주문에 대한 배달료를 계산하시오.
[배달료 기준표]
* 5km 미만: 2,000원
* 5km 이상 ~ 10km 미만: 4,000원
* 10km 이상 ~ 15km 미만: 6,000원
* 15km 이상 ~ 20km 미만: 8,000원

```python
# 배달료 계산 함수 정의
def delivery_fee(distance):
  if distance < 5:
    return 2000
  elif distance < 10:
    return 4000
  elif distance < 15:
    return 6000
  elif distance < 20:
    return 8000

df29['배달료'] = df29['거리'].apply(delivery_fee)
df29
```

## 2.  월별로 배달료의 총합을 집계하시오.

```python
df29['주문시간'] = pd.to_datetime(df29['주문시간'])
df29['주문월'] = df29['주문시간'].dt.month
total_delivery_fee = df29.groupby(by = '주문월')['배달료'].sum()
total_delivery_fee
```

## 3. 배달료가 가장 많이 발생한 월을 찾고, 그 월의 총 배달료를 정수로 구하시오.

```python
top_month = total_delivery_fee.idxmax()
top_delivery_fee = total_delivery_fee.max()
print("배달료가 가장 많이 발생한 월:", top_month, ", 그 월의 총 배달료:", top_delivery_fee)
```

# Section 30 주말, 평일 구분

```python
import pandas as pd
df30 = pd.read_csv("delivery_time.csv")
df30
```

## 1. 주말 주문 건수와 평일 주문 건수를 구하시오.

```python
df30['주문시간'] = pd.to_datetime(df30['주문시간'])
df30['요일'] = df30['주문시간'].dt.dayofweek

# 0~4 : 평일 / 5,6 : 주말
df30['주말'] = df30['요일'] >= 5

# 주말 & 평일 주문 건수 계산
weekdays = sum(~df30['주말'])
weekends = sum(df30['주말'])

print("주말 주문건수:", weekends, ", 평일 주문건수:", weekdays)
```

## 2. 주말 주문 건수와 평일 주문 건수의 차이를 절대값으로 구하고 정수형으로 구하시오.

```python
result = abs(weekdays - weekends)
print(int(result))
```
