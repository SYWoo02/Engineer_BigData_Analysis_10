```python
import pandas as pd
import numpy as np

df = pd.DataFrame({"메뉴": ['아메리카노', '카페라떼', '에스프레소', '카페모카', '바닐라라떼'],
                   "가격": [4500, 5000, 4000, 5900, 5300],
                   "칼로리": [10, 110, np.nan, 210, np.nan],
                   "원두": ['과테말라','브라질','과테말라',np.nan, np.nan]
                   })
df
```

## [데이터] Q1 ~ 10

### Q1: '칼로리' 컬럼의 결측치를 칼로리 데이터 중 최솟값으로 채우시오.

```python
df.isnull().sum()
```

```python
df['칼로리'] = df['칼로리'].fillna(df['칼로리'].min())
df
```

### Q2: '원두' 컬럼의 결측치를 원산지 데이터 중 최빈값으로 채우시오.

```python
f = df['원두'].mode()[0]
df['원두'] = df['원두'].fillna(f)
df
```

### Q3: 가격이 5,000원 이상인 데이터의 수를 구하시오.

```python
data = df['가격'] >= 5000
len(df[data])
```

### Q4: '이벤트가' 컬럼을 만들고 기존 가격에서 50% 할인된 가격을 채우시오.

```python
df['이벤특가'] = df['가격'] * 0.5
df
```

### Q5: '칼로리' 컬럼을 삭제하시오.

```python
df = df.drop('칼로리', axis=1)
df
```

### Q6: 위에서부터 3개의 행만 선택하시오. (loc 사용)
