### 라이브러리 및 데이터 불러기기

```python
import pandas as pd
import numpy as np
data = pd.read_csv("/kaggle/input/bigdatacertificationkr/basic1.csv")
```

### EDA

```python
data.head(5)
```

### f2가 0값인 데이터를 age를 기준으로 오름차순 정렬하기

```python
data2 = data[data['f2'] == 0]
data3 = data2.sort_values('age')
data3
```

### 앞에서 부터 20개의 데이터를 추출한 후,
### f1 결측치(최소값)를 채우기 전과 후의 분산 차이를 계산하기 (소수점 둘째 자리까지)

```python
data3 = data3.head(20)

# 'f1'칼럼의 결측치 채우기 전 'f1'칼럼의 분산
before_var = data3['f1'].var()
print(before_var)

# 'f1'칼럼의 결측치 채운 후 'f1'칼럼의 분산
min = data3['f1'].min()
data4 = data3['f1'].fillna(min)
after_var = data4.var()
print(after_var)
```

```python
# 최종 결과 출력
res = abs(before_var - after_var)
print(round(res,2))
```

## 풀이 (예시답안)

```python
import pandas as pd

# 데이터 불러오기
df = pd.read_csv("../input/bigdatacertificationkr/basic1.csv")

# f2가 0인 데이터 정렬(age 오름차순)
cond = (df['f2']==0)
df = df[cond].sort_values('age', ascending=True).reset_index(drop=True)

# 앞에서 부터 20개의 데이터 
df = df[:20]

# f1 결측치(최소값)를 채우기 전과 후의 분산
df_var1 = df['f1'].var()
df['f1'] = df['f1'].fillna(df['f1'].min())
df_var2 = df['f1'].var()

# 소수점 둘째자리까지 출력
print(round(df_var1 - df_var2, 2))

# 정답 : 38.44




# 추가 설명 판다스와 넘파이 기본값 비교

import numpy as np
print("numpy:", np.var(df['f1']))
print("pandas(ddof=0):",df['f1'].var(ddof=0))

print("pandas:",df['f1'].var())
print("numpy(ddof=1):", np.var(df['f1'], ddof=1))

```
