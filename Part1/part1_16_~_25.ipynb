# Section 16 이상치, IQR

```python
import pandas as pd
df16 = pd.read_csv("type1_data1.csv")
```

## 1. 'views' 컬럼의 1사분위수, 3사분위수 그리고 IQR을 계산하시오.

```python
Q1 = df16['views'].quantile(0.25)
Q3 = df16['views'].quantile(0.75)
IQR = Q3 - Q1
print(Q1, Q3, IQR)
```

## 2. 이상치 조건에 맞는 데이터를 찾으시오. (이상치는 1사분위수 - (IQR * 1.5) 보다 작은 값과 3사분위수 + (IQR * 1.5)보다 큰 값

```python
outlier = df16[(df16['views'] < Q1 - (IQR * 1.5)) | (df16['views'] > Q3 + (IQR * 1.5))]
outlier
```

## 3. 이상치 데이터의 'views' 컬럼 합을 정수로 구하시오.

```python
sum = outlier['views'].sum()
print(int(sum))
```

# Section 17 이상치, 소수점 있는 데이터 찾기, 표준 편차

```python
import pandas as pd
df17 = pd.read_csv("type1_data1.csv")
```

## 1. 'views' 컬럼의 표준편차를 구하시오.

```python
std = df17['views'].std()
std
```

## 2. 'age' 컬럼의 이상치 (소수점 나이, 음수 나이, 0살)을 제거하고, 'views' 컬럼의 표준편차를 구하시오.

```python
outlier_age = (df17['age'] < 0) | (df17['age'] == 0) | (df17['age'] != round(df17['age'],0))
df17_ = df17[~outlier_age]
df17_
```

```python
std2 = df17_['views'].std()
std2
```

## 3. 이상치 제거 전후의 'views' 컬럼의 표준편차를 더하여, 반올림 후 소수 둘째 자리까지 구하시오.

```python
result = std + std2
print(round(result,2))
```

# Section 18 데이터(행) 기준 평균값, 인덱싱

```python
import pandas as pd
df18 = pd.read_csv("type1_data2.csv")
df18.head()
```

## 1. index '2001' 데이터(행)에서 평균보다 큰 값의 개수를 구하시오.

```python
df18_2001 = df18[df18['year'] == 2001]
df18_2001 = df18_2001.drop(columns='year')
df18_2001
```

```python
# axis='columns' or axis=1 옵션으로 넣어야 행의 평균 출력
# index '2001' 데이터(행)에서 평균
mean_2001 = df18_2001.mean(axis=1).values[0]

# 평균보다 큰 값의 개수
count_above_mean = (df18_2001.iloc[0] > mean_2001).sum()
print(count_above_mean)
```

## 2. index '2003' 데이터(행)에서 평균보다 작은 값의 개수를 구하시오.

```python
df18_2003 = df18[df18['year'] == 2003]
df18_2003 = df18_2003.drop(columns='year')
df18_2003
```

```python
# index '2003' 데이터(행)에서 평균
mean_2003 = df18_2003.mean(axis=1).values[0]

# 평균보다 큰 값의 개수
count_above_mean2 = (df18_2003.iloc[0] < mean_2003).sum()
print(count_above_mean2)
```

## 3. 두 개수를 더하시오.

```python
res = count_above_mean + count_above_mean2
print(res)
```

# Section 19 결측치(뒤의 값으로 대체), 그룹합

```python
import pandas as pd
df19 = pd.read_csv("type1_data1.csv")
df19.head()
```

## 1. 결측치를 바로 뒤에 있는 값으로 대체하시오. (바로 뒤의 값도 결측치일 경우, 뒤에 있는 데이터 중 가장 가까운 값으로 대체)

```python
df19 = df19.bfill() # 바로 뒤의 값으로 대체 -> bfill() / 앞의 값으로 대체 -> ffill()
df19.isnull().sum()
```

## 2. 'city'와 'f2' 컬럼을 기준으로 그룹합을 계산하시오.

```python
df19_2 = df19.groupby(['city', 'f2']).sum(numeric_only=True).reset_index() # reset_index(): 인덱스를 열로 전환
df19_2
```

## 3. 'views' 값이 세 번째로 큰 city 이름을 구하시오.

```python
top = df19_2.sort_values('views',ascending=False) # 내림차순 -> ascending=False
print(top.iloc[2,0])
```

# Section 21 시간 간의 차이 계산(분), 필터링

```python
import pandas as pd
df21 = pd.read_csv("delivery_time.csv")
df21.head()
```

## 1. 예상 도착 시간보다 늦게 도착한 건수를 구하시오.

```python
df21['실제도착시간'] = pd.to_datetime(df21['실제도착시간'])
df21['예상도착시간'] = pd.to_datetime(df21['예상도착시간'])

# 지연시간 계산
diff = df21['실제도착시간'] - df21['예상도착시간']
diff.dt.total_seconds() / 60 #분
df21['지연시간'] = diff

# 예상 도착 시간보다 늦게 도착한 건수 구하기
# 예상 보다 늦게 도착한 경우 df21['지연시간'] > 0
cond1 = df21['지연시간'] > pd.Timedelta(0) # df21['지연시간']은 timedelta64 타입
df21_1 = df21[cond1]
df21_1
```

## 2. 이 중 거리가 7km 이상인 데이터의 수를 정수로 구하시오.

```python
df21_2 = df21_1[df21_1['거리'] >= 7]
print(len(df21_2))
```

# Section 22 시간 간의 차이 계산(분), 그룹핑

```python
import pandas as pd
df22 = pd.read_csv("delivery_time.csv")
df22.head()
```

## 1. 주문 시간과 실제 도착 시간의 차이를 분 단위로 계산하시오.

```python
df22['주문시간'] = pd.to_datetime(df22['주문시간'])
df22['실제도착시간'] = pd.to_datetime(df22['실제도착시간'])
diff22 = (df22['실제도착시간'] - df22['주문시간']).dt.total_seconds()/60
diff22
```

## 2. 앱 종류별로 평균 도착 시간(분)을 계산하시오.

```python
df22['주문~도착 총시간'] = diff22
df22_ = df22.groupby(by = '앱종류')
mean = df22_['주문~도착 총시간'].mean()
mean
```

## 3. 평균적으로 가장 빠른 앱 종류를 찾고, 해당 앱의 평균 도착 시간을 분으로 반올림하여 정수로 구하시오.

```python
print(round(mean.min()))
```

# Section 23 시간 간의 차이 계산(분), 비율

```python
import pandas as pd
df23 = pd.read_csv("delivery_time.csv")
df23.head()
```

## 1. 각 결제 종류별로 실제 도착 시간이 예상 도착 시간보다 늦은 주문의 비율을 계산하시오.

```python
df23['실제도착시간'] = pd.to_datetime(df23['실제도착시간'])
df23['예상도착시간'] = pd.to_datetime(df23['예상도착시간'])
diff23 = (df23['실제도착시간'] - df23['예상도착시간']).dt.total_seconds()/60
df23['지연시간'] = diff23

# 실제 도착 시간이 예상 도착 시간보다 늦은 경우, df23['지연시간'] > 0
df23['지연여부'] = df23['지연시간'] > 0
res = df23.groupby(by='결제종류')['지연여부'].mean()
print(res)
```

## 2. 비율 중 가장 큰 값을 반올림하여 소수 둘째 자리까지 구하시오.

```python
max = res.max()
print(round(max,2))
```

# Section 24 그룹핑, 값 찾기, 필터링

```python
import pandas as pd
df24 = pd.read_csv("delivery_time.csv")
df24.head()
```

## 1. 사용자별 주문 거리의 합계가 50km 이상인 사람들의 결제 방식을 구하시오.

```python
# 사용자별 총 주문거리 계산
df24_total_distance = df24.groupby(by = 'user')['거리'].sum()

# 사용자별 총 주문거리가 50km 이상 필터링
cond24_1 = df24_total_distance >= 50
df24_distance = df24_total_distance[cond24_1]

# 주문 거리의 합계가 50km 이상인 사용자들의 데이터 필터링
filtered_data = df24[df24['user'].isin(df24_distance.index)]
filtered_data

# 필터링한 데이터에서 결제종류 추출
payment = filtered_data['결제종류'].value_counts()
payment
```

## 2. 이 결제 방식 중 가장 빈도가 높은 수를 구하시오.

```python
max = payment.max()
max
```

# Section 25 시간 간의 차이 계산(일)

```python
import pandas as pd
df25 = pd.read_csv("delivery_time.csv")
df25.head()
```

## 1. 각 사용자별로 첫 주문과 마지막 주문 사이의 시간 간격을 일 단위로 계산하시오.

```python
# 각 사용자별로 첫 주문과 마지막 주문 사이의 시간 간격 계산 (일 단위)
df25['주문시간'] = pd.to_datetime(df25['주문시간'])
# 마지막 주문
max_order = df25.groupby(by = 'user')['주문시간'].max()
# 첫 주문
min_order = df25.groupby(by = 'user')['주문시간'].min()
df25_order_interval = (max_order - min_order).dt.days
df25_order_interval
```

## 2. 시간차가 0일인 사용자를 제외하고, 나머지 사용자들의 평균 시간 간격(일 단위)을 계산하시오.

```python
# 1일 이하(즉, 0일)인 주문은 제외
cond25_1 = df25_order_interval > 0
mean = df25_order_interval[cond25_1].mean()
print(mean)
```

## 3. 평균 시간 간격보다 긴 시간 간격을 가진 사용자의 수를 정수로 구하시오.

```python
cond25_2 = df25_order_interval > mean
count25 = df25_order_interval[cond25_2]
print(len(count25))
```
