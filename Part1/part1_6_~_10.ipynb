# Section 6 필터링, 분산

```python
import pandas as pd
df6 = pd.read_csv('type1_data1.csv')
df6.head()
```

## 1. 'f3' 컬럼이 gold면서 'f2' 컬럼이 2인 데이터를 찾으시오.

```python
df_6 = df6
cond1 = df_6['f3'] == 'gold'
cond2 = df_6['f2'] == 2
df_6 = df_6[cond1 & cond2]
df_6
```

## 2. 찾은 데이터에서 'f1' 컬럼의 분산을 구하시오. (반올림 후 소수 둘째 자리까지 계산)

```python
var = df_6['f1'].var()
print(round(var,2))
```

# Section 7 값 변경(연산), 필터링 절댓값

```python
import pandas as pd
df7 = pd.read_csv('type1_data1.csv')
```

## 1. 모든 나이(age)에 1을 더하시오.

```python
df_7 = df7
df_7['age'] = df_7['age'] + 1
df_7.head(5)
```

## 2. 20대의 'views' 평균과 30대의 'views' 평균의 차이의 절댓값을 구하시오. (반올림 후 소수 둘째 자리까지 계산)

```python
cond_20 = (df_7['age'] >= 20) & (df_7['age'] < 30.0)
cond_30 = (df_7['age'] >= 30) & (df_7['age'] < 40.0)

result = abs(df_7[cond_20]['views'].mean() - df_7[cond_30]['views'].mean())
print(round(result,2))
```

# Section 8 시계열 데이터, 필터링, 데이터 개수

```python
import pandas as pd
df8 = pd.read_csv('type1_data1.csv')
```

## 1. 'subscribed' 컬럼이 2024년 2월인 데이터를 찾으시오.

```python
df8['subscribed'] = pd.to_datetime(df8['subscribed'])
df8['Year'] = df8['subscribed'].dt.year
df8['Month'] = df8['subscribed'].dt.month
df8
```

```python
cond_2024 = df8['Year'] == 2024
cond_2 = df8['Month'] == 2
df_8 = df8[cond_2024 & cond_2]
df_8
```

## 2. 위에서 찾은 데이터 중 'f3' 컬럼이 gold인 데이터의 개수를 구하시오.

```python
cond_gold = df_8['f3'] == 'gold'
print(len(df_8[cond_gold]))
```

# Section 9 필터링, 카테고리, 최빈값

```python
import pandas as pd
df9 = pd.read_csv('type1_data1.csv')
```

## 1. 'views' 컬럼 값이 1000 이하인 데이터 (결측치 제외)를 찾으시오.

```python
cond = df9['views'] <= 1000
df_9 = df9[cond]
df_9
```

## 2. 앞에서 구한 데이터 중 'f4' 컬럼의 최빈값을 구하시오.

```python
freq = df_9['f4'].value_counts()
print(freq.index[0])
```

# Section 10 그룹핑, 최댓값, 정렬

```python
import pandas as pd
df10 = pd.read_csv('type1_data1.csv')
```

## 1. 결측치가 있는 행을 삭제하시오.

```python
df10 = df10.dropna()
df10
```

## 2. 결측치가 삭제된 데이터를 사용하여 지역별(city) 평균을 계산하시오.

```python
df_10 = df10.groupby(['city']).mean(numeric_only=True)
df_10
```

## 3. 앞에서 계산한 지역별 평균 데이터에서 'f2' 컬럼 값이 가장 큰 지역을 구하시오.

```python
max = df_10.sort_values('f2', ascending= False)
print(max.index[0])

# print(df_10['f2'].idxmax())
```
