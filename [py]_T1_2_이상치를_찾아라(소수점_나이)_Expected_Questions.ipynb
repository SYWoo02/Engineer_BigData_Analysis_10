## 라이브러리 및 데이터 불러오기

```python
import pandas as pd
import numpy as np
data = pd.read_csv("basic1.csv")
data.head(5)
```

```python
data.describe()
```

## 소수점 데이터 찾기 (age 컬럼)

```python
# 원래 데이터에서 해당 데이터의 내림한 값을 뺏을 때 0이 아니라면?
# => 양의 정수가 아닌 것 (float)
cond = (data['age'] - np.floor(data['age'])) != 0
wrong_age = data[cond]
wrong_age
```

## 이상치를 포함한 데이터 올림, 내림, 버림의 평균값 (age 칼럼)

```python
# 올림: np.ceil()
# 내림: np.floor()
# 버림: np.trunc()

# 올림
ceil = np.ceil(wrong_age['age'])

# 내림
floor = np.floor(wrong_age['age'])

# 버림
trunc = np.trunc(wrong_age['age'])

#결과 출력
print(ceil.mean(), floor.mean(), trunc.mean())
```

## 평균값 더한 다음 출력

```python
# 평균값 더하고 결과 출력
res = ceil.mean() + floor.mean() + trunc.mean()
print(res)
```
